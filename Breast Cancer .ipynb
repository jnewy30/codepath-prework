```python
import pandas as pd

# read in the data
data = pd.read_csv("data.csv")

# select the columns to bin
cols = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
        'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
        'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst',
        'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst',
        'fractal_dimension_worst']

# create the bin labels
bin_labels = ['Low', 'Medium', 'High']

# create a new dataframe to store the binned data
binned_data = pd.DataFrame()

# loop through each column and bin the data
for col in cols:
    binned_data[col] = pd.cut(data[col], bins=3, labels=bin_labels)
    
# add the id and diagnosis columns to the binned data
binned_data['id'] = data['id']
binned_data['diagnosis'] = data['diagnosis']

# view the first 5 rows of the binned data
print(binned_data.head())
```

```python
#PCA (Principal Component Anaylsis 1.4) (Method of Analysis #2)
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the data into a Pandas dataframe
data = pd.read_csv("data.csv")

# Select the columns to use in the analysis
cols_to_use = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
               'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
               'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst',
               'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst',
               'fractal_dimension_worst']

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[cols_to_use])

# Perform PCA
pca = PCA()
pca.fit(data_scaled)

# Plot the explained variance ratio
import matplotlib.pyplot as plt

plt.plot(range(1, len(cols_to_use)+1), pca.explained_variance_ratio_)
plt.xlabel('Principal component')
plt.ylabel('Explained variance ratio')
plt.show()
```

```python
import pandas as pd

# Read CSV file
df = pd.read_csv('cancer patient data sets.csv')

# Clean up column names
df.columns = df.columns.str.strip().str.lower().str.replace('_', ' ').str.replace('(', '').str.replace(')', '')

# Define a function to remove outliers from a column
def remove_outliers(column):
    q1, q3 = column.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return column[(column >= lower_bound) & (column <= upper_bound)]

# Remove outliers from the age column
df['age'] = remove_outliers(df['age'])

# Print the DataFrame with outliers removed
print(df)
```
